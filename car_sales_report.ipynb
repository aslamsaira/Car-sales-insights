```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return '{p:.1f}%\n(${v:d})'.format(p=pct, v=val)
    return my_autopct
```

```python
# Aggregate gender counts
gender_counts = df['Gender'].value_counts()
# Calculate the total annual income for each gender
gender_income = df.groupby('Gender')['Annual Income'].sum()
# Plot Show
plt.figure(figsize=(10, 10))
plt.pie(gender_counts, labels=gender_counts.index, autopct=make_autopct(gender_income))
plt.title('Gender Distribution with Total Annual Income')
plt.show()
```

```python
# Calculations
color_counts = df['Color'].value_counts()
# Calculate the total annual income for each gender
gender_income = df.groupby('Color')['Annual Income'].sum()
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return '{p:.1f}%\n(${v:d})'.format(p=pct, v=val)
    return my_autopct
# Plot Show
plt.figure(figsize=(10, 10))
plt.pie(color_counts, labels=color_counts.index, autopct=make_autopct(gender_income))
plt.title('Gender Distribution')
plt.show()
```

```python
df = pd.read_csv('/content/Car Sales.xlsx - car_data.csv')
```

```python
customer_income = df.groupby('Customer Name')['Annual Income'].sum()
```

```python
top_5_customers = customer_income.nlargest(5)
```

```python
plt.figure(figsize=(12, 8))
bars = plt.bar(top_5_customers.index, top_5_customers.values)
plt.xlabel('Customer Name')
plt.ylabel('Sum of Annual Income')
plt.title('Top 5 Customers by Annual Income')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Top 5 Companies - Chevrolet, Dodge, Ford, Volkswagen, Mitsubishi.
Prefered Engine - Double Overhead Camshaft
Prefered Transmission - Auto

Highest Sales Regions - Austin, Janesville, Scottsdale, Pasco & Aurora
